```python
# 1.chọn đối tượng trong ảnh
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')
bmg = data[800:1200, 1500:11000]
print(data.shape)

iio.imsave('apple.jpg', bmg)
plt.imshow(bmg)
plt.show()
```

```python
#2. Tịnh tiến đơn
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg', mode='F')
bdata = nd.shift(data, (50, -30))

plt.imshow(bdata)
plt.show()
```

```python
#3. Thay đổi kính thước ảnh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')
print("Original shape:", data.shape)
zoomed_data = nd.zoom(data, (3, 3, 1))
print("Zoomed shape:", zoomed_data.shape)

plt.imshow(zoomed_data)
plt.axis('off')  
plt.show()

```

```python
#4. Xoảy ảnh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')
print(data.shape)

d1 = nd.rotate(data, 50)
plt.imshow(d1)
plt.show()

d2 = nd.rotate(data, 50, reshape=False)
plt.imshow(d2)
plt.show()
```

```python
#5. Dilation và Erosion

import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('world_cup.jpg', mode='F')
print(data.shape)

d1 = nd.binary_dilation(data)
plt.imshow(d1)
plt.show()

d2 = nd.binary_dilation(data, iterations=3)
plt.imshow(d2)
plt.show()





```

```python
#6. coordinate
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
from skimage.color import rgb2gray
img_color = iio.imread('world_cup.jpg')
data = rgb2gray(img_color)

print(data.shape)
V, H = data.shape
M = np.indices((V, H))
d = 5
q = 2 * d * np.random.ranf(M.shape) - d
mp = (M + q).astype(int)
dl = nd.map_coordinates(data, mp)
plt.imshow(dl, cmap='gray')
plt.title('Hiệu ứng nhiễu tọa độ')
plt.show()

```

```python
#7. biến đổi chung
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt
from skimage.color import rgb2gray

def GeoFun(outcoord):
    a = 10 * np.sin(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.sin(outcoord[1] / 10.0) + outcoord[1]
    return a, b

data_color = iio.imread('world_cup.jpg')
data = rgb2gray(data_color)

d1 = nd.geometric_transform(data, GeoFun)
plt.imshow(d1, cmap='gray')
plt.show()
```
